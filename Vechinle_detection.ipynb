```python
import zipfile
import os

# Path to the ZIP file
zip_file_path = '/content/vehicledetection.zip'

# Directory where you want to extract the contents
extract_dir = '/content/data'

# Create the directory if it doesn't exist
os.makedirs(extract_dir, exist_ok=True)

# Extract the ZIP file
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall(extract_dir)

print("Extraction completed!")
```

```python
from ultralytics import YOLO

# Load a model
from PIL import Image
model = YOLO("/content/runs/detect/train2/weights/best.pt")  # pretrained YOLOv8n model

# Run batched inference on a list of images
results = model(["/content/data/VehiclesDetectionDataset/valid/images/0013d97a44c586c0_jpg.rf.QohOjYeZzQhUb4dxfAUr.jpg", "/content/data/VehiclesDetectionDataset/valid/images/0013d97a44c586c0_jpg.rf.c129e0f75e464334397783e12eeddd69.jpg"])  # return a list of Results objects

# Process results list
# Visualize the results
for i, r in enumerate(results):
    # Plot results image
    im_bgr = r.plot()  # BGR-order numpy array
    im_rgb = Image.fromarray(im_bgr[..., ::-1])  # RGB-order PIL image

    # Show results to screen (in supported envir onments)
    r.show()

    # Save results to disk
    r.save(filename=f"results{i}.jpg")
```
